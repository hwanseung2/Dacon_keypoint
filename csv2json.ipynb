```python
import numpy as np
import pandas as pd
import os
import time
import copy
import json
from collections import OrderedDict

from sklearn.model_selection import train_test_split
```

## 1. Train

```python
# Data directory
prefix_dir = '/content/drive/MyDrive/motion-challenge'

df = pd.read_csv(f'{prefix_dir}/train_df.csv')

# image file name numpy
imgs = df.iloc[:, 0].to_numpy()[:100]

# label numpy
motions = df.iloc[:, 1:][:100]

# remove _x._y
columns = motions.columns.to_list()[::2]
class_labels = [label.replace('_x', '').replace('_y', '') for label in columns]

I=[]
for idx,name in enumerate(imgs):
  ii = {}
  ii["file_names"] = name
  ii["height"] =1080
  ii["width"] = 1920
  ii["id"] = idx
  I.append(ii)

A=[]
for idx,motion in enumerate(motions.to_numpy()):
  ii = {}
  ii["segmentation"] = None

  a_keypoints = []
  x=[]
  y=[]
  for i in range(0, motion.shape[0], 2): # 0, 48, 2
        a_keypoints.extend((float(motion[i]), float(motion[i+1]),2))
        x.append(float(motion[i]))
        y.append(float(motion[i+1]))
  x=np.array(x)
  y=np.array(y)
  xmin = x.min()
  ymin = y.min()
  xmax = x.max()
  ymax = y.max()
  ii["keypoints"] = a_keypoints

  ii["num_keypoints"] = 24
  ii["area"] = None

  ii["iscrowd"] = 0
  ii["image_id"] = idx
  ii["bbox"] = [xmin,ymin, xmax-xmin,ymax-ymin]
  ii["category_id"] = 1

  ii["id"] = idx+5000

  A.append(ii)

C=[]
c={}
c["id"]=1
c["supercategory"]="person"
c["name"] = "person"
c["keypoints"] = class_labels
C.append(c)

json_data = OrderedDict()
json_data["images"]=I
json_data["annotations"]=A
json_data["categories"]=C
```

```python
with open('./train.json', 'w', encoding='utf-8') as make_file:
  json.dump(json_data, make_file, ensure_ascii=False, indent='\t')
```

```python
if not os.path.exists(prefix_dir+'/train'):
  os.makedirs(prefix_dir+'/train')
```

```python
import shutil

for i in imgs:
  shutil.copy(prefix_dir+'/train_imgs/'+i,prefix_dir+'/train/'+i) 
```
